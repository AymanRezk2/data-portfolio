```python
def summarize_columns(df):
    summary = []

    for col in df.columns:
        unique_values = df[col].dropna().unique()  # Ø§Ù„Ù‚ÙŠÙ… Ø§Ù„ÙØ±ÙŠØ¯Ø© Ø¨Ø¯ÙˆÙ† NaN
        summary.append({
            'Column Name': col,
            'Unique Count': df[col].nunique(dropna=True),
            'NaN Count': df[col].isna().sum(),
            'Unique Values': list(unique_values)
        })

    return pd.DataFrame(summary)
```

```python
summarize_columns(df)

```

```python

df['Gross Profit']          = df['Sales'] - df['Operating Costs / COGS']
df['Operating Profit']      = df['Gross Profit'] - df['Operating Expenses']


df['Gross Margin %']        = np.where(df['Sales'] == 0, 0, df['Gross Profit'] / df['Sales'])
df['Operating Margin %']    = np.where(df['Sales'] == 0, 0, df['Operating Profit'] / df['Sales'])
df['OpEx % of Sales']       = np.where(df['Sales'] == 0, 0, df['Operating Expenses'] / df['Sales'])


df['Gross Margin %']        = df['Gross Margin %'].round(4)
df['Operating Margin %']    = df['Operating Margin %'].round(4)
df['OpEx % of Sales']       = df['OpEx % of Sales'].round(4)
```
